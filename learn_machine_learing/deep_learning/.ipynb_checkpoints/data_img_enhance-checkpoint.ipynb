```python
# 展示图像数据增强 api 使用
"""
主要包括几个大类：
resize：缩放
crop：裁剪
flip：翻转
brightness & contrast：光照和对比度
"""
import numpy as np 
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
name = './img/gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
sess = tf.Session()
img_decoded_val = sess.run(img_decoded)

print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
"""
resize：缩放

tf.image.resize_area
tf.image.resize_bicubic
tf.image.resize_nearest_neightor
"""
# (224, 224, 3)
# (365, 600, 3)
name = './img/gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 224, 224, 3])

resize_img = tf.image.resize_bicubic(img_decoded, [448, 448])

sess = tf.Session()
img_decoded_val = sess.run(resize_img)
img_decoded_val = img_decoded_val.reshape((448, 448, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
"""
crop：裁剪
tf.image.pad_to_bounding_box
tf.image.crop_to_bounding_box
tf.random_crop
"""
name = './img/gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 224, 224, 3])

padded_img = tf.image.pad_to_bounding_box(
    img_decoded, 50, 100, 400, 400)

sess = tf.Session()
img_decoded_val = sess.run(padded_img)
img_decoded_val = img_decoded_val.reshape((400, 400, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
"""
flip：翻转
tf.image.flip_up_down
tf.image.flip_left_right
tf.image.random_flip_up_down
tf.image.random_flip_left_right
"""
name = './img/gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 224, 224, 3])

flipped_img = tf.image.flip_up_down(img_decoded)

sess = tf.Session()
img_decoded_val = sess.run(flipped_img)
img_decoded_val = img_decoded_val.reshape((224, 224, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
"""
brightness & contrast：光照和对比度
tf.image.adjust_brightness
tf.image.random_brightness
tf.image.adjust_contrast
tf.image.random_contrast
"""
name = './img/gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 224, 224, 3])

new_img = tf.image.adjust_brightness(img_decoded, 0.5)

sess = tf.Session()
img_decoded_val = sess.run(new_img)
img_decoded_val = img_decoded_val.reshape((224, 224, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

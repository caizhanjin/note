    + 一个内嵌函数引用其外部作作用域的变量,就会得到一个闭包. 需满足3点:
        1. 必须有一个内嵌函数
        2. 内嵌函数必须引用外部函数中的变量
        3. 外部函数的返回值必须是内嵌函数
    + 作用：1保持住当前的运行环境，2根据外部作用域的局部变量来得到不同的结果

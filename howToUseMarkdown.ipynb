## 一、为什么使用Markdown

### 1.1 Markdown是什么

Markdown是一种方便记忆、书写的纯文本标记语言，用户可以使用这些标记符号以最小的输入代价生成极富表现力的文档。Markdown的核心是Web发布所需的HTML代码的简写，用简单的语法表达我们的想法。大家现在看到的文档就是一份纯Markdown写的文档。

### 1.2 为什么我们要用Markdown

> * 项目的文档说明，wiki说明
> * 整理知识，学习笔记
> * 撰写或发布随想、论文：很多文字工作者也在用，因为无干扰和整洁
> * 代码支持，语法统一，可多平台使用，不用过多考虑格式方面问题
> * 语法简单，上手容易

``` javascript
/**
* nth element in the fibonacci series.
* @param n >= 0
* @return the nth element, >= 0.
*/
function fib(n) {
  var a = 1, b = 1;
  var tmp;
  while (--n >= 0) {
    tmp = a;
    a += b;
    b = tmp;
  }
  return a;
}

document.write(fib(10));
```
